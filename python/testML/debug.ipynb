```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np

# สมมติว่าข้อมูลกราฟของคุณมีลักษณะดังนี้
# graphs เป็น numpy array ที่มี shape (จำนวนกราฟ, จำนวนจุดในแต่ละกราฟ, จำนวนคุณลักษณะต่อจุด)
# ในตัวอย่างนี้, สมมติว่ามี 1 คุณลักษณะต่อจุดคือค่า y
# y_labels เป็นความหวานของน้ำตาลสำหรับแต่ละกราฟ

# ตัวอย่างข้อมูล
graphs = np.random.rand(1000, 100, 1)  # 100 กราฟ, แต่ละกราฟมี 10 จุด, 1 คุณลักษณะ (ค่า y)
y_labels = np.random.rand(1000, 1)  # ความหวานของน้ำตาลสำหรับแต่ละกราฟ

# สร้างโมเดล LSTM
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(100, 1)))  # 50 ยูนิต LSTM, จุดข้อมูล 10 จุด, 1 คุณลักษณะ
model.add(Dense(1))  # ชั้นเอาต์พุตมี 1 ยูนิตสำหรับทำนายความหวานของน้ำตาล

# คอมไพล์โมเดล
model.compile(optimizer='adam', loss='mean_squared_error')

# ฝึกโมเดล
model.fit(graphs, y_labels, epochs=100, batch_size=32)

```

```python
predictions = model.predict(y_labels)
```
